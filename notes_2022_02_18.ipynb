## Example: Taylor Series

Write a function called `exp_taylor` which takes input parameters `x` and `N` and returns the Nth partial sum of the Taylor series

$$
\sum_{n=0}^N \frac{x^n}{n!}
$$

```python
import scipy.special as sp

def exp_taylor(x,N):
    n = np.arange(0,N+1)
    return np.sum(x**n/sp.factorial(n))
```

```python
exp_taylor(1,100)
```

```python
n = np.arange(0,5)
n
```

```python
x = 1
x**n
```

```python
sp.factorial(n)
```

```python
x**n/sp.factorial(n)
```

## Plotting

```python
x = [0,2,3,5,8]
y = [2,1,5,-2,3]
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,1,50)
y = np.sin(np.pi*x)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,1,50)
for k in range(0,5):
    y = np.sin(k*np.pi*x)
    plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-10,10,100)
for a in np.arange(0.05,1,0.02):
    y = np.exp(-a*x**2)*np.cos(x)
    plt.plot(x,y)
plt.show()
```

Let's do a parametric plot:

\begin{align*}
x &= 16 \sin^3(t) \\
y &= 13 \cos(t) - 5 \cos(2t) - 2 \cos(3t) - \cos(4t)
\end{align*}

for $t \in [0,2\pi]$.

```python
t = np.linspace(0,2*np.pi,100)
x = 16*np.sin(t)**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)
for k in range(1,10):
    plt.plot(k*x,k*y,'r')
plt.show()
```
