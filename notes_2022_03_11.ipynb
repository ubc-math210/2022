## Example 1

Approximate the following integral with error less than 0.001.

$$
\int_0^1 \frac{dx}{1 + x^4}
$$

First, let's plot $y=f(x)$. Looking at the plot we can estimate the integral should be approximately 0.8.

```python
x = np.linspace(-2,2,50)
y = 1/(1 + x**4)
plt.plot(x,y)
plt.show()
```

We need $f''(x)$ to find $N$ for trapezoid rule. Compute:

$$
f'(x) = -\frac{4x^3}{(1 + x^4)^2}
\hspace{10mm}
f''(x) = \frac{32x^6 - 12x^2 (1+x^4)}{(1+x^4)^3}
$$

```python
x = np.linspace(0,1,50)
d2f = (32*x**6 - 12*x**2*(1 + x**4))/(1 + x**4)**3
plt.plot(x,d2f), plt.grid(True)
plt.show()
```

```python
np.max(np.abs(d2f))
```

```python
a = 0; b = 1; K2 = 2.4; E = 0.001;
N = np.sqrt((b - a)**3/(12*E)*K2)
N
```

```python
N = int(N) + 1
x = np.linspace(a,b,N+1)
y = 1/(1 + x**4)
spi.trapezoid(y,x)
```

```python
(b-a)/N*np.sum(y[1:] + y[:-1])/2
```

## Example 2

Approximate the following integral with error less than 0.001:

$$
\int_0^1 \frac{dx}{1 + e^x}
$$

$$
f'(x) = -\frac{e^x}{(1+e^x)^2}
\hspace{10mm}
f''(x) = \frac{2e^{2x} - (1+e^x)e^x}{(1+e^x)^3}
$$

```python
x = np.linspace(0,1,50)
y = (2*np.exp(2*x) - (1+np.exp(x))*np.exp(x))/(1+np.exp(x))**3
plt.plot(x,y), plt.grid(True)
plt.show()
```

```python
np.max(np.abs(y))
```

```python
a = 0; b = 1; K2 = 0.091; E = 0.001;
N = np.sqrt((b - a)**3/(12*E)*K2)
N
```

```python
N = int(N) + 1
x = np.linspace(a,b,N+1)
y = 1/(1 + np.exp(x))
spi.trapezoid(y,x)
```
