## Differential equations

A **differential equation** is an equation involving an unknown function $y(t)$ and its derivatives $y',y'',\dots$

The **order** of a differential equation is the highest order derivative in the equation.

For example, the logistic equation

$$
y' = y(1 - y)
$$

is a first order equation. And the pendulum equation

$$
\ddot{\theta} + \frac{g}{L} \sin \theta = 0
$$

is a second order equation (where $g=9.81$ is acceleration due to gravity, and $L$ is the length of the pendulum).

A first order equation is **separable** if it is of the form

$$
y' = f(y) g(t)
$$

In other words, we can separate the variables in the equation. We can integrate to solve separable equations

$$
\int \frac{dy}{f(y)} = \int g(t) dt
$$

### Example

Find all solutions of the equation $y' = -ty$. The equation is separable and so we compute

$$
y' = -ty \ \Rightarrow \int \frac{dy}{y} = -\int t dt
\ \Rightarrow \ln(y) = -\frac{t^2}{2} + C
\ \Rightarrow y(t) = k e^{-t^2/2}
$$

Let's check it:

$$
\frac{d}{dt} \left( k e^{-t^2/2} \right) = -t k e^{-t^2/2} = -ty
$$

Let's plot this family of solutions $ y(t) = k e^{-t^2/2}$ for $k=-5,\dots,5$. Note that $k = y(0)$.

```python
t = np.linspace(0,3,50)
for k in range(-5,6):
    y = k*np.exp(-t**2/2)
    plt.plot(t,y,'b')
plt.grid(True)
plt.show()
```

## Slope fields

Consider a first order differential equation

$$
y' = f(t,y)
$$

Note that $f(t,y)$ give us complete information about the slope of the solution $y(t)$ that passes through the point $(t,y)$. The **slope field** (aka direction field) plots a short line of slope $f(t,y)$ at a grid of points in the $ty$-plane.

```python
L = 0.2
t0 = 0; t1 = 6; N = 20; t = np.linspace(t0,t1,N)
y0 = -3; y1 = 3; M = 20; y = np.linspace(y0,y1,M);
f = lambda t,y: -t*y
for n in range(0,N):
    for m in range(0,M):
        slope = f(t[n],y[m])
        theta = np.arctan(slope)
        dt = L*np.cos(theta)
        dy = L*np.sin(theta)
        plt.plot([t[n],t[n]+dt],[y[m],y[m]+dy],'b')
plt.grid(True), plt.axis('equal')
plt.show()
```

```python
L = 0.4
t0 = 0; t1 = 4*np.pi; N = 20; t = np.linspace(t0,t1,N)
y0 = -2*np.pi; y1 = 2*np.pi; M = 20; y = np.linspace(y0,y1,M);
f = lambda t,y: np.sin(t) + np.cos(y)
for n in range(0,N):
    for m in range(0,M):
        slope = f(t[n],y[m])
        theta = np.arctan(slope)
        dt = L*np.cos(theta)
        dy = L*np.sin(theta)
        plt.plot([t[n],t[n]+dt],[y[m],y[m]+dy],'b')
plt.grid(True), plt.axis('equal')
plt.show()
```
